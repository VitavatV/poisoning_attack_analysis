```python
import pandas as pd
import numpy as np
```

```python
df_list = {}
for mode in ['uniform', 'normal', 'laplace', 'beta', 'triangular']:

    poisoned = 0
    csv_file = fr'results\ML101_linear_CL_batch\{mode}\poisoned_{poisoned}\poisoned_{poisoned}.csv'
    df_baseline_csv = pd.read_csv(csv_file)
    df_baseline_csv = df_baseline_csv[df_baseline_csv['batch']==-1].reset_index(drop=True)
    
    poisoned = 25
    csv_file = fr'results\ML101_linear_CL_batch\{mode}\poisoned_{poisoned}\poisoned_{poisoned}.csv'
    df_25_csv = pd.read_csv(csv_file)
    df_25_csv = df_25_csv[df_25_csv['batch']==-1].reset_index(drop=True)
    
    poisoned = 50
    csv_file = fr'results\ML101_linear_CL_batch\{mode}\poisoned_{poisoned}\poisoned_{poisoned}.csv'
    df_50_csv = pd.read_csv(csv_file)
    df_50_csv = df_50_csv[df_50_csv['batch']==-1].reset_index(drop=True)
    
    poisoned = 75
    csv_file = fr'results\ML101_linear_CL_batch\{mode}\poisoned_{poisoned}\poisoned_{poisoned}.csv'
    df_75_csv = pd.read_csv(csv_file)
    df_75_csv = df_75_csv[df_75_csv['batch']==-1].reset_index(drop=True)
    
    poisoned = 100
    csv_file = fr'results\ML101_linear_CL_batch\{mode}\poisoned_{poisoned}\poisoned_{poisoned}.csv'
    df_100_csv = pd.read_csv(csv_file)
    df_100_csv = df_100_csv[df_100_csv['batch']==-1].reset_index(drop=True)

    df_list[mode] = {
        'baseline': df_baseline_csv,
        '25': df_25_csv,
        '50': df_50_csv,
        '75': df_75_csv,
        '100': df_100_csv
    }
```

```python
for mode in ['uniform', 'normal', 'laplace', 'beta', 'triangular']:
    print("======================================")
    print(f"Mode: {mode}")
    df_baseline_csv = df_list[mode]['baseline']
    df_25_csv = df_list[mode]['25']
    df_50_csv = df_list[mode]['50']
    df_75_csv = df_list[mode]['75']
    df_100_csv = df_list[mode]['100']
        
    # Convergent Training Loss: final loss value
    convergent_loss = [
        df_baseline_csv['loss'].iloc[-1],
        df_25_csv['loss'].iloc[-1],
        df_50_csv['loss'].iloc[-1],
        df_75_csv['loss'].iloc[-1],
        df_100_csv['loss'].iloc[-1]
    ]

    # Global Model Accuracy on clean test data: final accuracy value
    final_accuracy = [
        df_baseline_csv['accuracy'].iloc[-1],
        df_25_csv['accuracy'].iloc[-1],
        df_50_csv['accuracy'].iloc[-1],
        df_75_csv['accuracy'].iloc[-1],
        df_100_csv['accuracy'].iloc[-1]
    ]

    # Time to Convergence: first epoch where accuracy >= threshold (e.g., 0.95)
    def get_time_to_convergence(df, accuracy_threshold = 0.95):
        convergence_idx = df[df['accuracy'] >= accuracy_threshold].index
        return convergence_idx[0] if len(convergence_idx) > 0 else None
        
    acc_thresh = 0.90
    time_to_convergence_90 = [
        get_time_to_convergence(df_baseline_csv, acc_thresh),
        get_time_to_convergence(df_25_csv, acc_thresh),
        get_time_to_convergence(df_50_csv, acc_thresh),
        get_time_to_convergence(df_75_csv, acc_thresh),
        get_time_to_convergence(df_100_csv, acc_thresh)
    ]
    acc_thresh = 0.80
    time_to_convergence_80 = [
        get_time_to_convergence(df_baseline_csv, acc_thresh),
        get_time_to_convergence(df_25_csv, acc_thresh),
        get_time_to_convergence(df_50_csv, acc_thresh),
        get_time_to_convergence(df_75_csv, acc_thresh),
        get_time_to_convergence(df_100_csv, acc_thresh)
    ]
    acc_thresh = 0.70
    time_to_convergence_70 = [
        get_time_to_convergence(df_baseline_csv, acc_thresh),
        get_time_to_convergence(df_25_csv, acc_thresh),
        get_time_to_convergence(df_50_csv, acc_thresh),
        get_time_to_convergence(df_75_csv, acc_thresh),
        get_time_to_convergence(df_100_csv, acc_thresh)
    ]
    acc_thresh = 0.60
    time_to_convergence_60 = [
        get_time_to_convergence(df_baseline_csv, acc_thresh),
        get_time_to_convergence(df_25_csv, acc_thresh),
        get_time_to_convergence(df_50_csv, acc_thresh),
        get_time_to_convergence(df_75_csv, acc_thresh),
        get_time_to_convergence(df_100_csv, acc_thresh)
    ]
    acc_thresh = 0.50
    time_to_convergence_50 = [
        get_time_to_convergence(df_baseline_csv, acc_thresh),
        get_time_to_convergence(df_25_csv, acc_thresh),
        get_time_to_convergence(df_50_csv, acc_thresh),
        get_time_to_convergence(df_75_csv, acc_thresh),
        get_time_to_convergence(df_100_csv, acc_thresh)
    ]

    # Robustness Score: accuracy degradation compared to baseline
    robustness_100 = [
        df_baseline_csv['accuracy'].iloc[-1] - df_baseline_csv['accuracy'].iloc[-1],
        df_baseline_csv['accuracy'].iloc[-1] - df_25_csv['accuracy'].iloc[-1],
        df_baseline_csv['accuracy'].iloc[-1] - df_50_csv['accuracy'].iloc[-1],
        df_baseline_csv['accuracy'].iloc[-1] - df_75_csv['accuracy'].iloc[-1],
        df_baseline_csv['accuracy'].iloc[-1] - df_100_csv['accuracy'].iloc[-1]
    ]

    print(f"Convergent Training Loss: {convergent_loss}")
    print(f"Final Accuracy: {final_accuracy}")
    print(f"Time to Convergence 90% (epoch): {time_to_convergence_90}")
    print(f"Time to Convergence 80% (epoch): {time_to_convergence_80}")
    print(f"Time to Convergence 70% (epoch): {time_to_convergence_70}")
    print(f"Time to Convergence 60% (epoch): {time_to_convergence_60}")
    print(f"Time to Convergence 50% (epoch): {time_to_convergence_50}")
    print(f"Robustness Score: {robustness_100}")
    
    # Prepare data for the table
    metrics_table = pd.DataFrame({
        'Poisoned %': ['0 (Baseline)', '25', '50', '75', '100'],
        'Final Accuracy': final_accuracy,
        'Convergent Loss': convergent_loss,
        'Robustness (vs Baseline)': robustness_100,
        'Time to 50%': time_to_convergence_50,
        'Time to 60%': time_to_convergence_60,
        'Time to 70%': time_to_convergence_70,
        'Time to 80%': time_to_convergence_80,
        'Time to 90%': time_to_convergence_90
    })

    # Convert numpy types to native Python types for display
    metrics_table = metrics_table.applymap(lambda x: x.item() if isinstance(x, (np.generic,)) else x)

    print(metrics_table)
```
