```python
import torch
import torch.nn as nn

model_list = []
method = 'FL'
for n in [1]:
    for l in [1]:
        for r in [0,1,2,]:
            for p in [0,10,20,30,40,50]:
                experiment_name = f'{method}_ModelN{n}L{l}_Batchsize1_rev{r}'
                poisoned = f'poisoned_{p}percent'
                model_info = {
                    'experiment': experiment_name,
                    'poisoned_ratio': poisoned,
                    'init_model': fr'D:\github\poisoning_attack_analysis\Experiment002\results\{experiment_name}\{poisoned}\model_init.pt',
                    'last_model': fr'D:\github\poisoning_attack_analysis\Experiment002\results\{experiment_name}\{poisoned}\model_last.pt'
                }
                model_list.append(model_info)


for model_pair in model_list:
    print()
    print(f"Experiment: {model_pair['experiment']}, Poisoned ratio: {model_pair['poisoned_ratio']}")
    init_model = model_pair['init_model']
    last_model = model_pair['last_model']

    if "N1L1" in model_pair['experiment']:
        class MNISTNet(nn.Module):
            def __init__(self):
                super().__init__()
                self.fc1 = nn.Linear(28*28, 1)
            def forward(self, x):
                x = self.fc1(x)
                x = torch.sigmoid(x)
                return x

    
    init_model_obj = torch.load(init_model, weights_only=False)
    last_model_obj = torch.load(last_model, weights_only=False)

    # Extract state dicts from model objects
    init_state_dict = init_model_obj.state_dict() if isinstance(init_model_obj, nn.Module) else init_model_obj
    last_state_dict = last_model_obj.state_dict() if isinstance(last_model_obj, nn.Module) else last_model_obj

    for key in init_state_dict.keys():
        print(f'Key: {key}')
        print (f'  Shape: {init_state_dict[key].shape}')
        print(f'  Init mean: {init_state_dict[key].mean().item():.6f},'\
            f' std: {init_state_dict[key].std(unbiased=False).item():.6f},'
            f' min: {init_state_dict[key].min().item():.6f},'
            f' max: {init_state_dict[key].max().item():.6f}')
        print(f'  Last mean: {last_state_dict[key].mean().item():.6f},'\
            f' std: {last_state_dict[key].std(unbiased=False).item():.6f},'\
            f' min: {last_state_dict[key].min().item():.6f},'
            f' max: {last_state_dict[key].max().item():.6f}')
```
